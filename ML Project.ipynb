### Fixing Outliers

```python
def drop_outlier(data,name):
    iqr = np.percentile(data[name],75)-np.percentile(data[name],25)
    
    data.drop(data[data[name] > (1.5*iqr + np.percentile(data[name],75))].index,inplace = True)
    data.drop(data[data[name] < (np.percentile(data[name],25) - iqr*1.5)].index,inplace = True)    
```

### Dataprocessing

### LinearRegression

```python
X = data.iloc[:,:-1]         # Features - All columns but last
y = data.iloc[:,-1]          # Target - Last Column
```

```python
from sklearn.metrics import r2_score
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2)
```

```python
from sklearn import linear_model
reg = linear_model.LinearRegression()
reg.fit(X_train,y_train)
```

```python
from sklearn.preprocessing import StandardScaler
stand= StandardScaler()
Fit = stand.fit(X_train)
Xtrain_scl = Fit.transform(X_train)
Xtest_scl = Fit.transform(X_test)
```

```python
# import linear regression models
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
lr=LinearRegression()
fit=lr.fit(Xtrain_scl,y_train)
score = lr.score(Xtest_scl,y_test)
print('predicted score is : {}'.format(score))
print('..................................')
y_predict = lr.predict(Xtest_scl)
print('mean_sqrd_error is ==',mean_squared_error(y_test,y_predict))
rms = np.sqrt(mean_squared_error(y_test,y_predict)) 
print('root mean squared error is == {}'.format(rms))
#print(r2_score(Xtest_scl,y_test))
```

### RandomForestRegressor

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn import metrics
rnd= RandomForestRegressor(ccp_alpha=0.0)
fit_rnd= rnd.fit(Xtrain_scl,y_train)
y_pred = rnd.predict(Xtest_scl)
#rnd= RandomForestRegressor(ccp_alpha=0.0)
# performance on train data
print('Performance on training data using RFR:',rnd.score(Xtrain_scl,y_train))
# performance on test data
print('Performance on testing data using RFR:',rnd.score(Xtest_scl,y_test))
print('mean_sqrd_error is==',mean_squared_error(y_test,rnd.predict(Xtest_scl)))
print('root_mean_squared error of is==',np.sqrt(mean_squared_error(y_test,rnd.predict(Xtest_scl))))
r2score = metrics.r2_score(y_test, y_pred)
acc_K=metrics.r2_score(y_test, y_pred)
print('r2score: ',acc_K)
```

#### The accuracy score of RandomForestRegressor is highest among linear, lasso, and ridge regression. Here the highest accuracy means it predicts the quality of concert by using training, which contains independent variables, and also it gives less error rate.
